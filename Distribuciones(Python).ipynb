```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Iniciemos viendo las distribuciones de probabilidad más frecuentes (funciones de densidad de probabilidad):

Distribución Uniforme continua: $$ f(x)= \frac{1}{b-a}\mathbb{1}_{(a,b)}(x) $$

```python
import scipy.stats
```

```python
from scipy.stats import uniform
```

```python
n=10000 #simular
start=0
width=10
data_uniform=uniform.rvs(size=n, loc=start, scale=width) #tamaño de la muestra, inicio y fin
```

```python
ax = sns.distplot(data_uniform,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la función
                  color='green',
                  hist_kws={"linewidth": 22,'alpha':1})
ax.set(xlabel='X ', ylabel="Frecuencia")
```

Distribución Normal: $$f(x|\mu,\sigma^2)= \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}\mathbb{1}_{\mathbb{R}}(x)$$

```python
from scipy.stats import norm
```

```python
n=10000
start = 5
width = 1
data_normal=norm.rvs(size=n,loc=start,scale=width)
```

```python
ax = sns.distplot(data_normal,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la función
                  color='gray',
                  hist_kws={"linewidth": 22,'alpha':1})
ax.set(xlabel='X ', ylabel="Frecuencia")
```

Distribución Gamma: $$f(x|\alpha,\beta)= \frac{\beta^{\alpha} x^{\alpha-1}e^{-\beta x}}{\Gamma(\alpha)} \mathbb{1}_{\mathbb{R^+}}(x)$$ para  $$x,\alpha, \beta > 0$$

```python
from scipy.stats import gamma
```

```python
n=10000
start = 5
width = 1

data_gamma=gamma.rvs(a=5,size=n) #notar qué parámetros acepta la función
```

```python
ax = sns.distplot(data_gamma,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la función
                  color='blue',
                  hist_kws={"linewidth": 22,'alpha':1}) 
ax.set(xlabel='X ', ylabel="Frecuencia") 
```

Distribución Exponencial: $$f(x)= \lambda e^{-\lambda x}\mathbb{1}_{\mathbb{R^+}}(x)$$

```python
from scipy.stats import expon
```

```python
n=10000
start = 5
width = 1

data_expon = expon.rvs(scale=width,loc=start,size=n)
```

```python
ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='red',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

Distribución Poisson: $$P(X=x)= e^{-\lambda} \frac{\lambda^{x}}{x!}\mathbb{1}_{{0,1,2 \dots}}(x)$$

```python
from scipy.stats import poisson
```

```python
n=10000
start = 1
width = 1

data_poisson = poisson.rvs(mu=start,size=n)
```

```python
ax = sns.distplot(data_poisson,
                  kde=True,
                  bins=50,
                  color='black',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

Además de la Poisson, hay otras dos distribuciones discretas importantes:

Distribución Binomial:  $$P(X=x)=  {n \choose x} p^x(1-p)^{n-x}\mathbb{1}_{{0,1,2 \dots}, n}(x) $$



```python
from scipy.stats import binom
```

```python
data_binomial = binom.rvs(n=40,p=0.1,size=10000)
```

```python
ax = sns.distplot(data_binomial,
                  kde=True,
                  bins=50,
                  color='yellow',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

Distribución Bernoulli: $$P(X=x)=  {1 \choose x} p^x(1-p)^{1-x}\mathbb{1}_{{0,1,2 \dots}, n}(x)=p^x(1-p)^{1-x}\mathbb{1}_{{0,1,2 \dots}, n}(x) $$

```python
from scipy.stats import bernoulli
```

```python
data_bernoulli = bernoulli.rvs(p=0.1,size=10000) #como casi siempre, se utilizan los parámetros de la función que corresponde
```

```python
ax = sns.distplot(data_bernoulli,
                  kde=True,
                  bins=50,
                  color='brown',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

Notar que la probabilidad es casi $0$, por lo que la distribución está sesgada a la izquierda.

Los procedimientos para graficar son muy parecidos: identificar los parámetros de la función, $size$ (es el número de intentos) y el código propio por distribución.
