```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Iniciemos viendo las distribuciones de probabilidad m치s frecuentes (funciones de densidad de probabilidad):

Distribuci칩n Uniforme continua: $$ f(x)= \frac{1}{b-a}\mathbb{1}_{(a,b)}(x) $$

```python
import scipy.stats
```

```python
from scipy.stats import uniform
```

```python
n=10000 #simular
start=0
width=10
data_uniform=uniform.rvs(size=n, loc=start, scale=width) #tama침o de la muestra, inicio y fin
```

```python
ax = sns.distplot(data_uniform,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la funci칩n
                  color='green',
                  hist_kws={"linewidth": 22,'alpha':1})
ax.set(xlabel='Distribuci칩n uniforme continua ', ylabel="Frecuencia")
```

Distribuci칩n Normal: $$f(x|\mu,\sigma^2)= \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}\mathbb{1}_{\mathbb{R}}(x)$$

```python
from scipy.stats import norm
```

```python
n=10000
start = 5
width = 1
data_normal=norm.rvs(size=n,loc=start,scale=width)
```

```python
ax = sns.distplot(data_normal,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la funci칩n
                  color='gray',
                  hist_kws={"linewidth": 22,'alpha':1})
ax.set(xlabel='Distribuci칩n normal ', ylabel="Frecuencia")
```

Distribuci칩n Gamma: $$f(x|\alpha,\beta)= \frac{\beta^{\alpha} x^{\alpha-1}e^{-\beta x}}{\Gamma(\alpha)} \mathbb{1}_{\mathbb{R^+}}(x)$$ para  $$x,\alpha, \beta > 0$$

```python
from scipy.stats import gamma
```

```python
n=10000
start = 5
width = 1

data_gamma=gamma.rvs(a=5,size=n) #notar qu칠 par치metros acepta la funci칩n
```

```python
ax = sns.distplot(data_gamma,
                  bins=100, #grupos de datos
                  kde=True, #kernel de la funci칩n
                  color='blue',
                  hist_kws={"linewidth": 22,'alpha':1}) 
ax.set(xlabel='Distribuci칩n Gamma ', ylabel="Frecuencia") 
```

Distribuci칩n Exponencial: $$f(x)= \lambda e^{-\lambda x}\mathbb{1}_{\mathbb{R^+}}(x)$$

```python
from scipy.stats import expon
```

```python
n=10000
start = 5
width = 1

data_expon = expon.rvs(scale=width,loc=start,size=n)
```

```python
ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='red',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Distribuci칩n exponencial', ylabel='Frequency')
```

Distribuci칩n Poisson: $$P(X=x)= e^{-\lambda} \frac{\lambda^{x}}{x!}\mathbb{1}_{{0,1,2 \dots}}(x)$$

```python
from scipy.stats import poisson
```

```python
n=10000
start = 1
width = 1

data_poisson = poisson.rvs(mu=start,size=n)
```

```python
ax = sns.distplot(data_poisson,
                  kde=False, #notar que aqu칤 interesa la frecuencia, en las anteriores (continua) se mostraban las probabilidades
                  bins=50,
                  color='black',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Distribuci칩n Poisson', ylabel='Frequency')
```

Adem치s de la Poisson, hay otras dos distribuciones discretas importantes:

Distribuci칩n Binomial:  $$P(X=x)=  {n \choose x} p^x(1-p)^{n-x}\mathbb{1}_{{0,1,2 \dots}, n}(x) $$



```python
from scipy.stats import binom
```

```python
data_binomial = binom.rvs(n=40,p=0.1,size=10000)
```

```python
ax = sns.distplot(data_binomial,
                  kde=False,
                  bins=50,
                  color='yellow',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Distribuci칩n binomial', ylabel='Frequency')
```

Distribuci칩n Bernoulli: $$P(X=x)=  {1 \choose x} p^x(1-p)^{1-x}\mathbb{1}_{{0,1,2 \dots}, n}(x)=p^x(1-p)^{1-x}\mathbb{1}_{{0,1,2 \dots}, n}(x) $$

```python
from scipy.stats import bernoulli
```

```python
data_bernoulli = bernoulli.rvs(p=0.7,size=10000) #como casi siempre, se utilizan los par치metros de la funci칩n que corresponde
```

```python
ax = sns.distplot(data_bernoulli,
                  kde=False,
                  bins=50,
                  color='brown',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Distribuci칩n Bernoulli', ylabel='Frequency')
```

Notar que la probabilidad es casi $0$, por lo que la distribuci칩n est치 sesgada a la izquierda.

Los procedimientos para graficar son muy parecidos: identificar los par치metros de la funci칩n, $size$ (es el n칰mero de intentos) y el c칩digo propio por distribuci칩n.
